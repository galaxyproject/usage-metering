```python
import os
import pandas as pd
```

```python
## change according to where you place numusers_all.csv file
usage_data = pd.read_csv('../numusers_all.csv')

usage_data.drop('Unnamed: 0',axis = 1,inplace = True)
```

```python
usage_individual_tool = usage_data.groupby(['tool_name_version']).sum()
usage_individual_tool.reset_index(inplace = True)


usage_individual_tool.sort_values(by = 'num_users',ascending = False,inplace = True)
usage_individual_tool

ten_most_popular_tools_users = usage_individual_tool.head(10)
```

```python
import matplotlib.pyplot as plt

most_popular_tool_names = ten_most_popular_tools_users.tool_name_version.values
most_popular_tool_names_users = ten_most_popular_tools_users.num_users.values
plt.title('Number of users (summed across all months) for the ten most popular tools in terms of total number of users')
plt.ylabel('Number of users')
plt.xlabel('Tool_name_version')
plt.xticks(rotation = 90)
plt.bar(most_popular_tool_names,most_popular_tool_names_users)
```

```python
total_cpu_time_all = pd.read_csv('../total_cpu_time_all.csv')
total_cpu_time_all.drop('Unnamed: 0',axis = 1,inplace = True)
for tool in most_popular_tool_names:
    tool_total_cpu_time = total_cpu_time_all.loc[total_cpu_time_all['tool_name_version'] == tool]
    tool_total_cpu_time.drop('tool_name_version',axis = 1,inplace = True)
    tool_name = tool.split('/')[0]
    filename = 'total_cpu_time_' + tool_name + '.csv'
    filepath = os.path.join(os.getcwd(),'resource_requirements_ten_most_popular_tools_by_users','total_cpu_time',filename)
    tool_total_cpu_time.to_csv(filepath,index = False)
    
```
