```python
import squarify
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/peiyuanp/Desktop/Galaxy_Summer2021/query_results/processed_toolid/totalmemory_all.csv',skiprows = 1,names = ['month_year','totalmem','tool_name_version'])

# drop all entries with NaN
df = df.dropna()

# remove all leading and trailing whitespaces in month_year
df['month_year'] = df['month_year'].str.strip()
df
```

```python
df2 = pd.read_csv('/Users/peiyuanp/Desktop/Galaxy_Summer2021/query_results/processed_toolid/avg_memory_all.csv',skiprows = 1,names = ['month_year','avgmemory','tool_name_version'])

# drop all entries with NaN
df2 = df2.dropna()

# remove all leading and trailing whitespaces in month_year
df2['month_year'] = df2['month_year'].str.strip()
```

```python
df4 = pd.read_csv('/Users/peiyuanp/Desktop/Galaxy_Summer2021/query_results/processed_toolid/avg_cpu_time_all.csv',skiprows = 1,names = ['month_year','avg_cpu_time','tool_name_version'])

# drop all entries with NaN
df4 = df4.dropna()

# remove all leading and trailing whitespaces in month_year
df4['month_year'] = df4['month_year'].str.strip()

```

```python
df3 = pd.read_csv('/Users/peiyuanp/Desktop/Galaxy_Summer2021/query_results/processed_toolid/total_cpu_time_all.csv',skiprows = 1,names = ['month_year','total_cpu_time','tool_name_version'])

# drop all entries with NaN
df3 = df3.dropna()

# remove all leading and trailing whitespaces in month_year
df3['month_year'] = df3['month_year'].str.strip()

# converts the entries in the total_cpu_time column to integers
df3['total_cpu_time'] = pd.to_numeric(df3['total_cpu_time'])
```

```python
squarify.plot(df['totalmem'])
plt.title('Treemap of totalmemory of tools grouped by month')
```

```python
df_aggregateSum = df.groupby('month_year')[['totalmem']].sum()
a = df_aggregateSum.index.get_level_values(0).tolist()
squarify.plot(sizes = df_aggregateSum['totalmem'],label = a)
plt.title('The sum of totalmemory across all tools for each month')
plt.axis('off')
plt.savefig('aggregate_totalmem.png')
```

```python
df_aggregateAvg2 = df2.groupby('month_year')[['avgmemory']].mean()
a = df_aggregateAvg2.index.get_level_values(0).tolist()
squarify.plot(sizes = df_aggregateAvg2['avgmemory'],label = a)
plt.title('The average of average memory across all tools for each month')
plt.axis('off')
plt.savefig('aggregate_avgmem.png')
```

```python
df_aggregateSum3 = df3.groupby('month_year')[['total_cpu_time']].sum()
a = df_aggregateSum3.index.get_level_values(0).tolist()

squarify.plot(sizes = df_aggregateSum3['total_cpu_time'],label = a)
plt.title('The sum of total cpu time across all tools for each month')
plt.axis('off')
plt.savefig('aggregate_totalcputime.png')
```

```python
df_aggregateAvg4 = df4.groupby('month_year')[['avg_cpu_time']].mean()
a = df_aggregateAvg4.index.get_level_values(0).tolist()

squarify.plot(sizes = df_aggregateAvg4['avg_cpu_time'],label = a)
plt.title('The average of average cpu time across all tools for each month')
plt.axis('off')
plt.savefig('aggregate_averagecputime.png')
```
